```python
from bs4 import BeautifulSoup as bs
from urllib import request
from requests import get
from os import path, listdir
from codecs import open as copen
from time import sleep


FILEPATH = 'index.html'
INFECTED_FP = '-iot_infected_device_stats.csv'
INFECTED_FP2 = '-iot_infected_device_stats-circl.csv'
EXPOSED_FP = '-iot_exposed_device_stats.csv'
EXPOSED_FP2 = '-iot_exposed_device_stats-circl.csv'

if not path.isfile(FILEPATH):
	f = request.urlopen('https://cra.circl.lu/opendata/variot/iot-exposed-infected-device-stats/')
	html = f.read().decode('utf-8')
	f = copen(FILEPATH, 'w', 'utf-8')
	f.write(html)
	f.close()
	print('Getting original data.')

f = copen(FILEPATH, 'r', 'utf-8')
soup = bs(f, 'html.parser')
file_links = [l['href'][:-1] for l in soup.find_all('a', href=True)[6:]][:-1]

for fl in file_links:
	if not path.isfile('./data/infected/{fl}-infected.csv'):
		r = get(f'https://cra.circl.lu/opendata/variot/iot-exposed-infected-device-stats/{fl}/{fl}{INFECTED_FP}')
		if r.status_code != 404:
			f = copen(f'./data/infected/{fl}-infected.csv', 'w', 'utf-8')
			f.write(r.text)
			f.close()
		else:
			r = get(f'https://cra.circl.lu/opendata/variot/iot-exposed-infected-device-stats/{fl}/{fl}{INFECTED_FP2}')
			if r.status_code != 404:
				f = copen(f'./data/infected/{fl}-infected.csv', 'w', 'utf-8')
				f.write(r.text)
				f.close()
			else:
				f = copen(f'./data/infected/{fl}-infected.csv', 'w', 'utf-8')
				f.write('infection,geo,country,count\n')
				f.close()
		print(f'GET {fl}{INFECTED_FP} {r.status_code}')
		sleep(.25)
	if not path.isfile('./data/exposed/{fl}-exposed.csv'):
		r = get(f'https://cra.circl.lu/opendata/variot/iot-exposed-infected-device-stats/{fl}/{fl}{EXPOSED_FP}')
		if r.status_code != 404:
			f = copen(f'./data/exposed/{fl}-exposed.csv', 'w', 'utf-8')
			f.write(r.text)
			f.close()
		else:
			r = get(f'https://cra.circl.lu/opendata/variot/iot-exposed-infected-device-stats/{fl}/{fl}{EXPOSED_FP2}')
			if r.status_code != 404:
				f = copen(f'./data/exposed/{fl}-exposed.csv', 'w', 'utf-8')
				f.write(r.text)
				f.close()
			else:
				f = copen(f'./data/exposed/{fl}-exposed.csv', 'w', 'utf-8')
				f.write('vendor,model,type,geo,country,count,possible_vulnerability\n')
				f.close()
		print(f'GET {fl}{EXPOSED_FP} {r.status_code}')
		sleep(.25)
```
